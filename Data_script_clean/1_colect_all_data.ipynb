# Data collection

```python
# import df_final_web_data_pt

import pandas as pd

dft1 = pd.read_csv('df_final_web_data_pt_1.txt', engine= 'python')
dft1

```

```python
# import df_final_web_data_pt2

dft2 = pd.read_csv('df_final_web_data_pt_2.txt', engine= 'python')
dft2
```

```python
# concat dft1 with dft2


web_data = pd.concat([dft1, dft2])
web_data
```

```python
# import df_final_demo

dft_demo = pd.read_csv('df_final_demo.txt', engine= 'python')
dft_demo
```

```python
#import df_final_experiment_clients.

dft_exp = pd.read_csv('df_final_experiment_clients.txt', engine= 'python')
dft_exp
```

# Data wrangler

```python
# merge dft_exp with dft_demo

data_merge1 = pd.merge(dft_exp, dft_demo, on='client_id')
data_merge1.head()
```

```python
data_merge2 = pd.merge(data_merge1, web_data, on='client_id')
data_merge2.head()
```

```python
data_merge2.gendr.unique()
```

```python
data_merge2= data_merge2.drop(data_merge2[data_merge2['gendr'] == "X"].index)
```

```python
data_merge2.gendr.value_counts()
```

```python
data_merge2.isnull().sum()
```

```python
data_merge3 = data_merge2.dropna(subset=['clnt_tenure_yr'])
```

```python
data_merge3.isnull().sum()
```

```python
data_merge4 = data_merge2.dropna(subset=['clnt_age'])
```

**DataMeta**
