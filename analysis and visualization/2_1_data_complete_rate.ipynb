# Gold Control Customers
<span style="font-size: 30px;">

Steps breakdown by visit ID and not Client ID


</span>


```python
import pandas as pd


# Inicializar um dicionário para armazenar as contagens
count_dict = {}

# Percorrer o DataFrame
for index, row in gold_c_control.iterrows():
    steps = row['process_step'].split()  # Separar as palavras da coluna process_step
    current_count = 1  # Inicializar a contagem
    for step in steps:
        key = (row['visit_id'], step)  # Criar a chave para o dicionário
        if key in count_dict:  # Verificar se a chave já existe no dicionário
            count_dict[key] += current_count  # Incrementar a contagem
        else:
            count_dict[key] = current_count  # Adicionar a chave ao dicionário com a contagem inicial
        current_count += 1  # Incrementar a contagem para a próxima iteração

# Criar um novo DataFrame a partir do dicionário
df_gold_vi_s = pd.DataFrame(list(count_dict.items()), columns=['key', 'count'])
df_gold_vi_s[['visit_id', 'process_step']] = pd.DataFrame(df_gold_vi_s['key'].tolist(), index=df_gold_vi_s.index)  # Separar a chave em colunas "client_id" e "process_step"
df_gold_vi_s = df_gold_vi_s.pivot(index='visit_id', columns='process_step', values='count').reset_index()  # Transformar dados em formato de tabela
df_gold_vi_s = df_gold_vi_s.fillna(0)  # Preencher valores nulos com 0


df_gold_vi_s
```

```python
df_gold_vi_s['confirm'].sum()
```

```python
df_gold_vi_s['confirm'].describe()
```

**Gold / control - Condition for Complete and Error** 

```python
#Whenever
"""
s1 < s
s2 < s1
s3 < s2
s3 < confirm
and confirm != 0
== success

else error

also if start = 0 == null
"""

import pandas as pd

def verifica_completude(row):
    if row['step_1'] <= row['start'] and row['step_2'] <= row['step_1'] and row['step_3'] <= row['step_2'] and row['confirm'] <= row['step_3']:
        return row['confirm']
    elif row['start'] > row['confirm']:
        return 1
    else:
        return None

def substituir_nulos(row):
    if pd.isnull(row['complete']) and row['confirm'] <= row['start'] and row['confirm'] != 0:
        return row['confirm']
    else:
        return row['complete']

df_gold_vi_s['complete'] = df_gold_vi_s.apply(verifica_completude, axis=1)
df_gold_vi_s['complete'] = df_gold_vi_s.apply(substituir_nulos, axis=1)
df_gold_vi_s['Error'] = df_gold_vi_s['complete'].apply(lambda x: 1 if pd.isnull(x) or x == 1 else 0)
df_gold_vi_s.head(30)
```

**Correction of Conditions** 

```python
# if Start is bigger the confirm there is a error
import numpy as np

df_gold_vi_s['Error'] = np.where(df_gold_vi_s['start'] > df_gold_vi_s['confirm'], 1, df_gold_vi_s['Error'])
df_gold_vi_s.tail(30)
```

```python
import numpy as np

df_gold_vi_s['Error'] = np.where((df_gold_vi_s['confirm'] == df_gold_vi_s['start']) & 
                                (df_gold_vi_s['start'] == df_gold_vi_s['step_1']) & 
                                (df_gold_vi_s['step_1'] == df_gold_vi_s['step_2']) & 
                                (df_gold_vi_s['step_2'] == df_gold_vi_s['step_3']), 0, df_gold_vi_s['Error'])
df_gold_vi_s.tail(5)
```

```python
gold_c_control_new = gold_c_control.drop(['confirm', 'start', 'step_1', 'step_2', 'step_3' ], axis=1)
gold_c_control_new
```

```python

gold_c_control_new = pd.merge(gold_c_control_new, df_gold_vi_s, on='visit_id')
gold_c_control_new
```
