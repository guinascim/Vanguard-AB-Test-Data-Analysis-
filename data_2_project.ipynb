```python
import pandas as pd


with open('df_final_web_data_pt_1.txt', 'r') as file:
  
    lines = file.readlines()

  
    data1 = []

  
    for line in lines:
        
        words = line.strip().split(',') 

 
        data1.append(words)


df1 = pd.DataFrame(data1)
df1
```

```python
with open('df_final_web_data_pt_2.txt', 'r') as file:
  
    lines = file.readlines()

  
    data2 = []

  
    for line in lines:
        
        words = line.strip().split(',') 

 
        data2.append(words)


df2 = pd.DataFrame(data1)
df2
```

```python
web_data = pd.concat([df1, df2])
web_data
```

```python
novo_dataframe = web_data.drop(web_data.index[0])
novo_dataframe
```
